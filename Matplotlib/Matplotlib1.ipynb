```python
df = pd.read_csv('batsman_season_record.csv')
```

```python
df
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2,color='yellow')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='blue')
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2,color='red')
plt.title('Runs scored by batsman in 2015, 2016, 2017')
plt.xticks(np.arange(df.shape[0]),df['batsman'],rotation=0)
plt.xlabel('Batsman name')
plt.ylabel('Runs scored')
```

```python
# stack bar chart
plt.bar(df['batsman'],df['2017'],color='blue',width=0.8,label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],color='green',width=0.8,label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2017']+df['2016']),color='yellow',width=0.8,label='2015')
plt.legend()
```
