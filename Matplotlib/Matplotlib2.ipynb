```python
batters = pd.read_csv('batter.csv')
batters.shape
```

Annotations

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
plt.xlabel('Avg')
plt.ylabel('Strike rate')
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

# 3D Plot

3D Scatter plot

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter3D(batters['runs'],batters['avg'],batters['strike_rate'])
ax.set_title('IPL batsman analysis')
ax.set_xlabel('Runs')
ax.set_ylabel('Avg')
ax.set_zlabel('Strike Rate')
```

3D Line plot

```python
x = [0,1,5]
y = [0,10,13]
z = [0,13,20]

fig = plt.figure()
ax = plt.subplot(projection='3d')
ax.scatter3D(x,y,z,s=[100,100,100])
ax.plot3D(x,y,z,color='red')
```

3D Surface Plots : 
Loss function -> How much the result of a machine learning algorithm is correct or not


```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.log(yy)
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
plt.show()
```

Contour plot : plots 3D graph into 2D

```python
xx, yy = np.meshgrid(x,y)
z = xx**2 + yy**2
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()
p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
ax = plt.subplot()
p = ax.contourf(xx, yy, z, camp = 'viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.sin(yy)
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()
p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
plt.show()
```

# Heatmap

How values are distributed

```python
delivery = pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
delivery.head()
```

```python
temp_df = delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run']==6)]
```

```python
grid = temp_df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
grid
```

```python
plt.figure(figsize=(20,10))
plt.imshow(grid)
plt.yticks(delivery['overs'].unique(),list(range(1,21)))
plt.xticks(np.arange(0,6),list(range(1,7)))
plt.colorbar()
```
